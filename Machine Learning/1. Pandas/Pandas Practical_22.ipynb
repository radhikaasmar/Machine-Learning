# Pandas Pivot Table

```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/student_results.csv')
df
```

```python
df.pivot_table(index='Class')
```

```python
df.pivot_table(index='Class', columns = 'Study hrs')
```

```python
df.pivot_table(index='Class', columns = 'Study hrs', aggfunc='count')
```

```python
df.pivot_table(index='Class', columns = 'Study hrs', aggfunc='max')
```

```python
df.pivot_table(index='Class', columns = 'Study hrs', aggfunc='sum')
```

```python
df.pivot_table(index='Class', columns = 'Study hrs', fill_value=0)
```

```python
df.pivot_table(index='Student ID', columns = 'Class', fill_value=0, margins=True)
```

```python
df.pivot_table(index='Student ID', columns = 'Class', fill_value=0, margins=True, aggfunc='sum')
```
