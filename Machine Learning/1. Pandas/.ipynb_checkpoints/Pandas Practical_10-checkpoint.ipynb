# Handling Missing Values in Python #3

```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv')
df = df.head()
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how = 'any')
```

```python
df.dropna(axis=1, how = 'any')
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='all')
```

```python
df
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv')
df = df.head()
df
```

```python
df.dropna(how = 'all' )
```

```python
df.dropna(thresh=4)
```

```python
df.dropna(subset=['ID'])
```

```python
df.dropna(subset=['Name'])
```
