# Handling Missing Values in Python

```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv')
df
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv', na_values=['not available', 'no values'])
df
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv')
df
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv', na_values={'Companies': 'no values'})
df
```
