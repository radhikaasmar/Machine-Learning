# Pandas fillna()

```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/Fortune_10.csv')
df = df.head()
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Name':'none', 'Industry':0, 'Growth':0})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill', axis=1)
```

```python
df.fillna(method='bfill', axis=1)
```

```python
df.fillna(0,limit=1)
```

```python
df.fillna(method='ffill',limit=1)
```

```python
df.fillna(5,inplace=True)
df
```
