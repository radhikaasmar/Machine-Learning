# Pandas Melt()

```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/student_results.csv')
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars=['Student ID'])
```

```python
pd.melt(df, id_vars=['Student ID'], value_vars=['Class'])
```

```python
pd.melt(df, id_vars=['Student ID'], value_vars=['Study hrs'])
```

```python
pd.melt(df, id_vars=['Student ID'], value_vars=['Study hrs'], var_name='Category')
```

```python
pd.melt(df, id_vars=['Student ID'], value_vars=['Study hrs'], var_name='Category', value_name='Data')
```
