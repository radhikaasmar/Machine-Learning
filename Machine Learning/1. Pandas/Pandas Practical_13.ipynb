# Pandas interpolate()

```python
import pandas as pd
```

```python
df = pd.read_csv('Downloads/student_results.csv')
df
```

```python
df.interpolate(method='linear')
```

```python
df.interpolate(method='time')
```

```python
df.interpolate(method='index')
```

```python
df.interpolate(method='nearest')
```

```python
df.interpolate(method='polynomial', order=1)
```

```python
df.interpolate(method='polynomial', order=2)
```

```python
df.interpolate(method='spline', order=2)
```

```python
df.interpolate()
```

```python
df.interpolate(limit=2)
```

```python
df.interpolate(limit_direction='both', limit=1)
```

```python
df.interpolate(limit_direction='both', limit=2)
```

```python
df.interpolate(limit_area='inside')
```

```python
df.interpolate(limit_area='outside')
```
