# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
plt.pie([1])
plt.show()
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
```

```python
plt.pie(class1_students, labels=classes)
plt.show()
```

```python
explode = [0,0,0,0,0,]
plt.pie(class1_students, labels=classes, explode =explode)
plt.show()
```

```python
explode = [0,0,0.2,0,0,]
plt.pie(class1_students, labels=classes, explode =explode)
plt.show()
```

```python
colors = ['c','b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode =explode, colors=colors)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True, radius=1.4)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True, radius=1.4, startangle =180)
plt.show()
```

```python
textprops = {'fontsize':15}

plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True, radius=1.4, startangle =180,
       textprops=textprops)
plt.show()
```

```python
wedgeprops = {'linewidth':4, 'width':1, 'edgecolor':'k'}

plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True, radius=1.4, startangle =180,
       textprops=textprops, wedgeprops=wedgeprops)
plt.show()
```

```python
plt.figure(figsize=(5,5))

plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True, radius=1.4, startangle =180,
       textprops=textprops, wedgeprops=wedgeprops)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode =explode, colors=colors,
       autopct='%0.2f%%', shadow=True, radius=1.4, startangle =180,
       textprops=textprops, wedgeprops=wedgeprops)
plt.legend()
plt.show()
```
