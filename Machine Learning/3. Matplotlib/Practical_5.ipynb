# Matplotlib Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of 'Indian Artificial Intelligence Production (IAIP) Class"
 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students)
```

```python
plt.barh(classes, class1_students)
```

```python
plt.bar(classes, class1_students, width=0.2, align='edge', color='y', 
        edgecolor='m', linewidth=5, alpha=0.9, linestyle='--',
       label='Class 1 Students', visible=False)
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))

plt.bar(classes, class1_students, width=0.6, align='edge', 
        edgecolor='b', linewidth=5, alpha=0.9, linestyle='--',
       label='Class 1 Students')
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))

plt.title('Bar Chart of IAIP Class', fontsize=30)
plt.xlabel("Classes", fontsize=25)
plt.ylabel('No. of Students', fontsize=25)

plt.barh(classes, class1_students,  align='edge', 
        edgecolor='b', linewidth=5, alpha=0.9, linestyle='--',
       label='Class 1 Students')
```

```python
plt.figure(figsize=(16,9))

plt.title('Bar Chart of IAIP Class', fontsize=30)
plt.xlabel("Classes", fontsize=25)
plt.ylabel('No. of Students', fontsize=25)

plt.barh(classes, class1_students,  align='edge', 
        edgecolor='b', linewidth=5, alpha=0.9, linestyle='--',
       label='Class 1 Students')

plt.barh(classes, class2_students,  align='edge', color='g',
        edgecolor='b', linewidth=5, alpha=0.9, linestyle='--',
       label='Class 2 Students')
```

```python
plt.figure(figsize=(16,9))

plt.title('Bar Chart of IAIP Class', fontsize=30)
plt.xlabel("Classes", fontsize=25)
plt.ylabel('No. of Students', fontsize=25)

classes_index = np.arange(len(classes))
width=0.2

plt.bar(classes_index, class1_students,    width,
       label='Class 1 Students')

plt.bar(classes_index+width, class2_students, width,  color='g',
        
       label='Class 2 Students')
```

```python
plt.figure(figsize=(16,9))

plt.title('Bar Chart of IAIP Class', fontsize=30)
plt.xlabel("Classes", fontsize=25)
plt.ylabel('No. of Students', fontsize=25)

classes_index = np.arange(len(classes))
width=0.2

plt.xticks(classes_index+width, classes)

plt.bar(classes_index, class1_students,    width,
       label='Class 1 Students')

plt.bar(classes_index+width, class2_students, width,  color='g',
        
       label='Class 2 Students')
```

```python
plt.figure(figsize=(16,9))

plt.title('Bar Chart of IAIP Class', fontsize=30)
plt.xlabel("Classes", fontsize=25)
plt.ylabel('No. of Students', fontsize=25)

classes_index = np.arange(len(classes))
width=0.2

plt.xticks(classes_index+width, classes)

plt.bar(classes_index, class1_students,    width,
       label='Class 1 Students')

plt.bar(classes_index+width, class2_students, width,  color='g',
        
       label='Class 2 Students')
plt.bar(classes_index+width+width, class3_students, width,  color='y',
        
       label='Class 2 Students')
```

```python
plt.figure(figsize=(16,9))

plt.title('Bar Chart of IAIP Class', fontsize=30)
plt.ylabel("Classes", fontsize=25)
plt.xlabel('No. of Students', fontsize=25)

classes_index = np.arange(len(classes))
width=0.2

plt.yticks(classes_index+width, classes)

plt.barh(classes_index, class1_students,    width,
label='Class 1 Students')

plt.barh(classes_index+width, class2_students, width,  color='g', label='Class 2 Students')
plt.barh(classes_index+width+width, class3_students, width,  color='y', label='Class 2 Students')

plt.legend()
```
