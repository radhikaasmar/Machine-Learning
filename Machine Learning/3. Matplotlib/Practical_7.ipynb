# Matplotlib Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv')
df.shape
```

```python
df = pd.read_csv('googleplaystore.csv', nrows=1000)
df.shape
```

```python
x = df['Rating']
y = df['Reviews']
```

```python
plt.scatter(x,y)

plt.title('Google PLay Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y, c='r', marker='*', s=100)

plt.title('Google PLay Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y, c='r', marker='*', s=100, linewidth=10,
           edgecolors='g')

plt.title('Google PLay Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y, c='r', marker='*', s=100, linewidth=10,
           edgecolors='g')

plt.scatter(x,df['Installs'], c='r', marker='*', s=100, linewidth=10, color='b',
           edgecolors='g')

plt.title('Google PLay Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
```
