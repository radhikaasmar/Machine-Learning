# Matplotlib Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)

plt.title("ML Students age histograms")
plt.xlabel('Students age category')
plt.ylabel('No. Students age')

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist(ml_students_age, bins, rwidth=0.9, histtype='bar', orientation='horizontal', label='ML Student')

plt.title("ML Students age histograms")
plt.xlabel('Students age category')
plt.ylabel('No. Students age')
plt.legend()

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist(ml_students_age, bins, rwidth=0.9, histtype='bar', label='ML Student')
plt.hist(py_students_age, bins, rwidth=0.9, histtype='bar', label='PY Student', color='g')

plt.title("ML Students age histograms")
plt.xlabel('Students age category')
plt.ylabel('No. Students age')
plt.legend()

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist([ml_students_age, py_students_age], bins, rwidth=0.9, histtype='bar', label=['ML Student', 'PY Student'], color=['b','g'])

plt.title("ML Students age histograms")
plt.xlabel('Students age category')
plt.ylabel('No. Students age')
plt.legend()

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.figure(figsize=(16,9))
plt.hist([ml_students_age, py_students_age], bins, rwidth=0.9, histtype='bar', label=['ML Student', 'PY Student'], color=['b','g'])

plt.title("ML Students age histograms")
plt.xlabel('Students age category')
plt.ylabel('No. Students age')
plt.legend()

plt.show()
```
