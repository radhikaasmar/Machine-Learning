# Seaborn Heatmap

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
globalWarming_df = pd.read_csv('Downloads/Who_is_responsible_for_global_warming.csv')
globalWarming_df
```

```python
globalWarming_df = globalWarming_df.drop(columns = ['Country Code', 'Indicator Name', 'Indicator Code'],
                      axis=1).set_index('Country Name')
globalWarming_df = globalWarming_df.dropna()
globalWarming_df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df, vmin=0, vmax=21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df, vmin=0, vmax=21, cmap='coolwarm')
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df, vmin=0, vmax=21, cmap='coolwarm', 
            annot=True)
```

# Seaborn Heatmap Part 2

```python
plt.figure(figsize=(16,9))

annot_kws = { 'fontsize': 10,
             'fontstyle': 'italic',
             'color':'k',
             'alpha':0.9,
             'rotation':'vertical',
             'verticalalignment': 'center',
             'backgroundcolor': 'w'
}
                    
sns.heatmap(globalWarming_df, vmin=0, vmax=21, cmap='coolwarm', 
           annot=True, annot_kws = annot_kws)
```

```python
plt.figure(figsize=(16,9))
                    
sns.heatmap(globalWarming_df, vmin=0, vmax=21, cmap='coolwarm', 
           annot=True, linewidth=4)
```

```python
plt.figure(figsize=(16,9))
                    
sns.heatmap(globalWarming_df, vmin=0, vmax=21, cmap='coolwarm', 
           annot=True, linewidth=4, linecolor='k')
```

```python
plt.figure(figsize=(16,9))
                    
sns.heatmap(globalWarming_df, cbar=False)
```

```python
plt.figure(figsize=(16,9))
                    
sns.heatmap(globalWarming_df, cbar=False, xticklabels=False, yticklabels=False)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }                   
sns.heatmap(globalWarming_df, cbar_kws=cbar_kws)
```

```python
plt.figure(figsize=(16,9))
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']                   
sns.heatmap(globalWarming_df, xticklabels=np.arange(0,15), 
            yticklabels=country_code)
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(globalWarming_df)
ax.set(title="Heatmap",
      xlabel="Years",
      ylabel="Country Name",)
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(globalWarming_df)
ax.set(title="Heatmap",
      xlabel="Years",
      ylabel="Country Name",)
sns.set(font_scale=1)
```

# Seaborn Part 3

```python
globalWarming_df.corr()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df.corr(), annot=True, linewidth=3)
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(globalWarming_df.corr(), annot=True, linewidth=3)
ax.tick_params(size=10, color='k', labelsize=10, labelcolor='k')
plt.title("Heatmap", fontsize=25)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                         columns = np.append(cancer_dataset['feature_names'], ['target']
                        ))
cancer_df
```

```python
plt.figure(figsize = (25,25))
ax = sns.heatmap(cancer_df.corr(), annot=True, linewidth=3)
ax.tick_params(size=10, color='k', labelsize=10, labelcolor='k')
plt.title("Heatmap of Breast Cancer Patients", fontsize=25)
plt.show()
```
