# Seaborn Histogram using distplot

```python
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.distplot(tips_df['size'])
```

```python
sns.distplot(tips_df['tip'])
```

```python
sns.distplot(tips_df['total_bill'])
```

```python
sns.distplot(tips_df['total_bill'], bins=10)
```

```python
sns.distplot(tips_df['total_bill'], bins=100)
```

```python
sns.distplot(tips_df['total_bill'], bins=100, hist=False)
```

```python
sns.distplot(tips_df['total_bill'],  kde=False)
```

```python
sns.distplot(tips_df['total_bill'],  rug=True)
```

```python
sns.distplot(tips_df['total_bill'],  fit=norm, kde=False)
```

```python
sns.distplot(tips_df['total_bill'],  color='r')
```

```python
sns.distplot(tips_df['total_bill'],  vertical=True)
```

```python
sns.distplot(tips_df['total_bill'],  norm_hist=True)
```

```python
sns.distplot(tips_df['total_bill'],  axlabel='Total Bill')
```

```python
sns.distplot(tips_df['total_bill'],  label='Total Bill')
plt.legend()

plt.title('Histrogram of Total Bill')
```

```python
plt.figure(figsize=(16,9))

sns.set()

sns.distplot(tips_df['total_bill'],  label='Total Bill')
plt.legend()

plt.title('Histrogram of Total Bill')
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(16,9))

sns.set()

sns.distplot(tips_df['total_bill'],  bins=bins)
plt.xticks(bins)

plt.title('Histrogram of Total Bill)
plt.show()
```

```python
plt.figure(figsize=(16,9))

sns.set()

sns.distplot(tips_df['total_bill'],  
            hist_kws={'color':'red', 
                      'edgecolor':'white', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))

sns.set()

sns.distplot(tips_df['total_bill'],  
            hist_kws={'color':'red', 
                      'edgecolor':'white', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9},
            kde_kws={'color':'yellow', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9},
            rug=True,
            rug_kws={'color':'blue', 
                      'linewidth':3, 
                      'linestyle':'--',
                     'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))

sns.set()

sns.distplot(tips_df['total_bill'],  
            hist_kws={'color':'red', 
                      'edgecolor':'white', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9},
            kde=False,
             fit=norm,
            fit_kws = {'color':'yellow', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9} ,
            kde_kws={'color':'yellow', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9},
            rug=True,
            rug_kws={'color':'blue', 
                      'linewidth':3, 
                      'linestyle':'--',
                     'alpha':0.9})
```

```python
plt.figure(figsize=(16,9))

sns.set()

sns.distplot(tips_df['total_bill'],  
            hist_kws={'color':'red', 
                      'edgecolor':'white', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9},
            kde=False,
             fit=norm,
            fit_kws = {'color':'pink', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9} ,
            kde_kws={'color':'yellow', 
                      'linewidth':5, 
                      'linestyle':'--',
                     'alpha':0.9},
            rug=True,
            rug_kws={'color':'blue', 
                      'linewidth':3, 
                      'linestyle':'--',
                     'alpha':0.9},
            label='TB')

plt.title("Histogram of Restaurant Total Bill", fontsize =20)
plt.xlabel('Total Bill', fontsize=15)
plt.legend()
plt.show()
```
