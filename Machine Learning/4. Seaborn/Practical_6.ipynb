# Seaborn Barplot

```python
import seaborn as sns
import numpy as np
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.barplot()
```

```python
sns.barplot(x=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day ,y=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day ,y=tips_df.total_bill, hue=tips_df.sex)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df)
```

```python
order = ['Sun', 'Thur', 'Fri', 'Sat']
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, order=order)
```

```python
hue_order = ['Female', 'Male']
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, hue_order=hue_order)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           estimator=np.mean)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           estimator=np.max)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           ci=12)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           estimator=np.max, n_boot=2)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           orient='h')
```

```python
sns.barplot(x='total_bill',y='day' , hue='sex', data=tips_df)
```

```python
sns.barplot(x='total_bill', y='size', hue='sex', data=tips_df, orient='h')
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           color='r')
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           palette='Blues')
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           saturation=0.3)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           errcolor='0.5')
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
           errwidth=12)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, 
          capsize=0.3)
```

```python
sns.barplot(x='day' ,y='total_bill', hue='sex', data=tips_df, dodge=False)
```

```python
kwargs={'alpha':0.9, 'linestyle': ':', 'linewidth': 5, 'edgecolor':'k'}

sns.barplot(x='day' ,y='total_bill', data=tips_df, **kwargs)
```

```python
sns.barplot(x='day' ,y='total_bill', data=tips_df, alpha=0.9, linestyle='--', linewidth=3, edgecolor='g')
```

```python
ax = sns.barplot(x='day' ,y='total_bill', data=tips_df, alpha=0.9, 
                 linestyle='--', linewidth=3, edgecolor='g')

ax.set(title = 'Barplot of Tips Dataframe',
      xlabel='Days',
      ylabel='Tota; Bill')
```
